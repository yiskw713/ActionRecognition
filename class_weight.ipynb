```python
import glob
import os
import pandas as pd
import torch
```

```python
df_train = pd.read_csv('./utils/train.csv')
df_val = pd.read_csv('./utils/val.csv')
```

```python
df = pd.concat([df_train, df_val])
```

```python
df.sort_values('class_id')
```

```python
class_nums = {}
for i in range(400):
    class_nums[i] = 0;
```

```python
for i in range(len(df)):
    class_nums[df.iloc[i, 2]] += 1
```

```python
class_nums
```

```python
class_num = []
for val in class_nums.values():
     class_num.append(val)
```

```python
class_num = torch.tensor(class_num)
```

```python
class_num
```

```python
total = class_num.sum().item()

frequency = class_num.float() / total
median = torch.median(frequency)

class_weight = median / frequency
```
